How to iterate over rows?

```python
for index, row in df.iterrows():
    print(row['Fare'], row['Age'])
```

```python
for index, i in df.iterrows():
    print(int(i['Fare']))
```

```python
for index, row in df.iterrows():
    print(int(row['Fare']) * 100)
```

```python
sum  = 0
for index, row in df.iterrows():
    sum = sum + (int(row['Fare']))
print(sum)
```
